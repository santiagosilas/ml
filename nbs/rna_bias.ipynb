```python
# Exemplo para entendimento do BIAS
# Considere um Perceptron com apenas uma entrada
# Input                Output
#             w_i
# x_1 ------------------> sigmoide(w_i * x_i)
```

```python
from matplotlib import pyplot as plt
import numpy as np 
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
# Valores possíveis para o input
x_1 = np.arange(-5, 5, 0.1)
```

```python
# mudar o peso altera  stepness da sigmoide
# Mas se fosse desejável que a rede imprimisse zero quando peso == 2.0 ?

w_1 = 0.5
plt.plot(x_1, sigmoid(x_1 * w_1), label='w_1={0}'.format(w_1)) # peso = 0.5

w_1 = 1.0
plt.plot(x_1, sigmoid(x_1 * w_1) , label='w_1={0}'.format(w_1)) # peso = 1.0

w_1 = 2.0
plt.plot(x_1, sigmoid(x_1 * w_1), label='w_1={0}'.format(w_1)) # peso = 2.0

plt.title('Função Sigmoide')
plt.xlabel('Valores possíveis para x_1')

plt.legend(loc=4)


plt.show()
```

```python
 # O Bias permite isso!
# Input                Output
#             w_i
# x_i ------------------> sigmoide(w_i * x_i)
# 
w_1 = 1.0
x_0, bias = 1.0, -0.5
plt.plot(x_1, sigmoid(x_1 * w_1 + x_0 * bias), label='w_1={0}, bias (w_0) = {1}'.format(w_1, bias))

x_0, bias = 1.0, 0.0
plt.plot(x_1, sigmoid(x_1 * w_1 + x_0 * bias), label='w_1={0}, bias (w_0) = {1}'.format(w_1, bias))

x_0, bias = 1.0, 0.5
plt.plot(x_1, sigmoid(x_1 * w_1 + x_0 * bias), label='w_1={0}, bias (w_0) = {1}'.format(w_1, bias))

plt.title('Função Sigmoide')
plt.xlabel('Valores possíveis para x_1')

plt.legend(loc=4)

plt.show()
```
